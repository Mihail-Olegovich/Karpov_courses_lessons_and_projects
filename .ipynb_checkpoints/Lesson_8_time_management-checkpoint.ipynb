Посмотрим на данные за период с апреля по декабрь.

```python
bikes_q2_sample_apr = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q2_sample_apr.csv')
bikes_q2_sample_may = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q2_sample_may.csv')
bikes_q2_sample_jun = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q2_sample_jun.csv')
bikes_q3_sample_july = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q3_sample_july.csv')
bikes_q3_sample_aug = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q3_sample_aug.csv')
bikes_q3_sample_sep = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q3_sample_sep.csv')
bikes_q4_sample_oct = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q4_sample_oct.csv')
bikes_q4_sample_nov = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q4_sample_nov.csv')
bikes_q4_sample_dec = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/all_data/bikes_q4_sample_dec.csv')
```

```python
bikes = pd.concat([bikes_q2_sample_apr, bikes_q2_sample_may, bikes_q2_sample_jun, bikes_q3_sample_july,\
                   bikes_q3_sample_aug, bikes_q3_sample_sep, bikes_q4_sample_oct, bikes_q4_sample_nov,\
                   bikes_q4_sample_dec])
```

```python
bikes['start_time'] = pd.to_datetime(bikes['start_time'])
```

```python
bikes_by_usertype = bikes.groupby(['usertype', 'start_time']).agg({'trip_id': 'count'})\
        .rename(columns={'trip_id': 'count'})\
        .unstack(level=0).resample(rule='D').sum()
```

```python
bikes_by_usertype
```

```python
bikes_by_usertype = bikes_by_usertype.T.reset_index(drop=True).T #reset_index для колонок
```

```python
bikes_by_usertype.rename(columns={0: 'Customer', 1: 'Subscriber'}, inplace=True)
```

Сделайте преобразование по дням для каждой группы пользователей (usertype), затем выберите дни, в которые число аренд, сделанных customers, было больше, чем у subscribers.

```python
bikes_by_usertype.query('Customer > Subscriber')
```

Отлично! Следующий шаг – более подробно посмотреть на летний период.

Еще один плюс использования дат в качестве индексов – возможность выбрать данные за интересующий нас промежуток времени. В переменную bikes_summer сохраните наблюдения с 1 июня по 31 августа. Затем запишите в top_destination наиболее популярный пункт назначения (его название). Агрегируйте данные по дням и определите, в какой день в полученный пункт (top_destination) было совершено меньше всего поездок. Дату сохраните в bad_day, отформатировав timestamp с помощью .strftime('%Y-%m-%d').bikes

```python
bikes = bikes.set_index('start_time')
bikes_summer = bikes['2018-06-01': '2018-08-31']
```

```python
bikes_summer
```

```python
top_destination = bikes_summer.groupby('to_station_name').count().idxmax()[0]
```

```python
bad_day = bikes_summer.query('to_station_name == @top_destination')\
        .resample(rule='D').count()\
        .idxmin()[0].strftime('%Y-%m-%d')
```
