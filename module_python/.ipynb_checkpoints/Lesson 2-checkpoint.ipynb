```python
min_pickups = taxi.groupby('borough').pickups.sum().idxmin()
```

```python
min_pickups
```

```python
taxi.groupby(['borough', 'hday']).agg({'pickups': 'mean'})
```

```python
pickups_by_mon_bor = taxi.groupby(['pickup_month', 'borough'], as_index=False)\
                        .pickups.sum()\
                        .sort_values(by='pickups', ascending=False)
```

```python
def temp_to_celcius(f_temp: pd.Series) -> pd.Series:
        return ((f_temp - 32) * 5.0) / 9.0 #векторизованные вычисления
```

```python
taxi['temp'][:5]
```
