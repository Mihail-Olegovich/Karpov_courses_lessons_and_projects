```python
trans_by_trans = transaction_data.transaction.value_counts()
fig, ax = plt.subplots(figsize=(7, 5))
sns.barplot(x=trans_by_trans.index, y=trans_by_trans.values)

ax.set_ylabel('Количество')
ax.set_xlabel('Тип завершения')
ax.set_title('Распределение по типам транзакций')
ax.bar_label(ax.containers[0])
```

Распределение кол-ва успешных транзакций на человека

```python
success_per_user = transaction_data.query(' transaction == "successfull" ') \
        .groupby('name', as_index=False) \
        .agg({'transaction': 'count'})
```

```python
sns.displot(data=success_per_user, x='transaction')
```

```python
transaction_data_update\
        .groupby(['minute', 'name'], as_index=False)\
        .agg({'transaction': 'count'})\
        .pivot(index='minute', columns='name', values='transaction')\
        .fillna(0)
```

Создание графика количества транзакций по минутам от начала дня

```python
transaction_data_update['true_minute'] \
        = transaction_data_update.date.dt.minute + transaction_data_update.date.dt.hour * 60
```

```python
transaction_data_update
```

```python
plt.figure(figsize=(25,15))
sns.barplot(data = transaction_data_update\
        .groupby(['true_minute', 'name'], as_index=False)\
        .agg({'transaction': 'count'})\
        .pivot_table('transaction', index='true_minute', aggfunc='sum')\
        .fillna(0).reset_index()\
        .rename(columns={'transaction': 'transactions_number'}),\
        x='true_minute', y='transactions_number', color='Blue', alpha=0.5)
```

Дополнительный проект

```python
prod_activations_logs = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_courses_lessons_and_projects/module_python/all_data/prod_activations_logs.csv', sep=';')
tm_sales_1 = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_courses_lessons_and_projects/module_python/all_data/tm_sales_1.csv', sep=';')
tm_sales_2 = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_courses_lessons_and_projects/module_python/all_data/tm_sales_2.csv', sep=';')
tm_sales_3 = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_courses_lessons_and_projects/module_python/all_data/tm_sales_3.csv', sep=';')
```

```python
prod_activations_logs
```

```python
tm_sales = pd.concat([tm_sales_1, tm_sales_2, tm_sales_3], ignore_index=True)
```

```python
tm_sales = tm_sales.dropna(subset=['SUBS_ID'])
```

```python
tm_sales['SUBS_ID'] = tm_sales['SUBS_ID'].apply(lambda x: 'id' + x if not x.startswith('id') else x)
```

```python
tm_sales
```

```python
prod_activations_logs['START_DTTM'] = pd.to_datetime(prod_activations_logs['START_DTTM'], format='%d-%m-%Y %H:%M')
prod_activations_logs['END_DTTM'] = pd.to_datetime(prod_activations_logs['END_DTTM'], format='%d-%m-%Y %H:%M')
```

```python
prod_activations_logs['difference'] = prod_activations_logs.END_DTTM - prod_activations_logs.START_DTTM
```

```python
prod_activations_logs = prod_activations_logs[prod_activations_logs.difference >= '5 minutes' ]
```

```python
tm_sales.merge(prod_activations_logs, how='inner', on=['SUBS_ID', 'PROD_ID'])\
        .sort_values('SUBS_ID')
```
