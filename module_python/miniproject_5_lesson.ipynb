```python
transaction_data_update\
        .groupby(['minute', 'name'], as_index=False)\
        .agg({'transaction': 'count'})\
        .pivot(index='minute', columns='name', values='transaction')\
        .fillna(0)
```

```python
transaction_data_update['true_minute'] =\
            (transaction_data_update.date - transaction_data_update.date.dt.round('D'))
```

```python
transaction_data_update.true_minute = transaction_data_update.true_minute.apply(lambda x: x.total_seconds()//60.0)
```

```python
transaction_data_update
```

```python
plt.figure(figsize=(25,15))
sns.barplot(data = transaction_data_update\
        .groupby(['true_minute', 'name'], as_index=False)\
        .agg({'transaction': 'count'})\
        .pivot_table('transaction', index='true_minute', aggfunc='sum')\
        .fillna(0).reset_index()\
        .rename(columns={'transaction': 'transactions_number'}),\
        x='true_minute', y='transactions_number', color='Blue', alpha=0.5)
```

```python
prod_activations_logs = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/prod_activations_logs.csv', sep=';')
tm_sales_1 = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/tm_sales_1.csv', sep=';')
tm_sales_2 = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/tm_sales_2.csv', sep=';')
tm_sales_3 = pd.read_csv('C:/Users/1/JupyterProjects/Karpov_curses_lessons_and_projects/tm_sales_3.csv', sep=';')
```

```python
prod_activations_logs
```

```python
tm_sales = pd.concat([tm_sales_1, tm_sales_2, tm_sales_3], ignore_index=True)
```

```python
tm_sales = tm_sales.dropna(subset=['SUBS_ID'])
```

```python
tm_sales['SUBS_ID'] = tm_sales['SUBS_ID'].apply(lambda x: 'id' + x if not x.startswith('id') else x)
```

```python
tm_sales
```

```python
prod_activations_logs['START_DTTM'] = pd.to_datetime(prod_activations_logs['START_DTTM'], format='%d-%m-%Y %H:%M')
prod_activations_logs['END_DTTM'] = pd.to_datetime(prod_activations_logs['END_DTTM'], format='%d-%m-%Y %H:%M')
```

```python
prod_activations_logs['difference'] = prod_activations_logs.END_DTTM - prod_activations_logs.START_DTTM
```

```python
prod_activations_logs = prod_activations_logs[prod_activations_logs.difference >= '5 minutes' ]
```

```python
tm_sales.merge(prod_activations_logs, how='inner', on=['SUBS_ID', 'PROD_ID'])\
        .sort_values('SUBS_ID')
```
